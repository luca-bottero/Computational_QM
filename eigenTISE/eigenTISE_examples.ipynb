## Examples of 1D potentials

We explore the time-independent solution of some QM potentials.

We begin with the simple harmonic oscillator, where $$\large V_{SHO}(x) = \omega^2x^2$$

```python
import numpy as np
import matplotlib.pyplot as plt

import importlib as imp
import eigenTISE
imp.reload(eigenTISE)
from eigenTISE import TISE_1D

x = np.linspace(-10., 10., 1000)

def V_SHO(x):
    return x**2

SHO = TISE_1D(x, V_SHO)
SHO.solve()

SHO.plot_eigenvalues()
plt.show()
plt.plot(SHO.eigenfunctions[0]**2)
plt.show()

normalization_array = np.array([np.trapz(wf**2,x) for wf in SHO.eigenfunctions])

plt.plot(normalization_array)
```

The double well potential is: $$\large V_{DWP}(x) = x^4 - x^2$$

```python
def V_DWP(x):
    return x**4 - x**2

x = np.linspace(-1.35,1.35,1000)

plt.plot(x, V_DWP(x))

DWP = TISE_1D(x, V_DWP)

DWP.solve()
DWP.plot_eigenvalues()

plt.show()
for i in range(4):
    plt.plot(x, DWP.eigenfunctions[i]**2)
    plt.title(str(i))
    plt.show()

normalization_array = np.array([np.trapz(wf**2,x) for wf in DWP.eigenfunctions])

plt.plot(normalization_array)
```
