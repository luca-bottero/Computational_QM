```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
x = np.linspace(-5, 5, 1000)
wavefunction = np.exp(1j * x) #* np.exp(-0.1*x**2)

# Get the magnitude and phase
magnitude = np.abs(wavefunction)
phase = np.angle(wavefunction)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the wavefunction
ax.plot(x, magnitude * np.cos(phase), magnitude * np.sin(phase), label='Wavefunction')

# Plot the x-axis
ax.plot(x, np.zeros_like(x), np.zeros_like(x), label='X-axis', color='black')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Re[Psi]')
ax.set_zlabel('Im[Psi]')
ax.legend()

# Enable interactive mode
plt.ion()
plt.show()
```
